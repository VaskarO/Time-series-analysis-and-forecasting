#### Augmented Dickey-Fuller Test for Stationarity 

##### The Augmented Dickey-Fuller (ADF) test is a statistical test commonly used to check for stationarity in a time series. It tests the null hypothesis that a unit root is present in the series, which indicates non-stationarity.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
raw_data = pd.read_csv('./datasets/TSLA.csv')
raw_data.head()
```

```python
df = raw_data.loc[:,['Date', 'Close']]
```

```python
plt.plot(df['Close'])
```

```python
X= df['Close'].values
results = adfuller(X)
print("ADF Test statistics : ", results)
print(f'P-value : {results[1]}')
print('Critical values:')
for key, value in results[4].items():
    print('\t%s: %.3f' % (key, value))

if results[0]<results[4]["5%"]:
    print("Reject Null Hypotheisis (H0) : Time series is stationary.")
else:
    print('Failed to Reject Null Hypothesis')
```
